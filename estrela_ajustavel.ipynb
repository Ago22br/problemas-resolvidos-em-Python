```python
#Resultado Com Interface

import matplotlib.pyplot as plt

print("Estrela Com tamanho Ajustavel")
print("Estrela tamanho padrão 8x8.")
escala = int(input("Escolha um tamanho: ")) #para alterar o tamanho basta mudar trocar o valor numerico

#variaveis que definem o tamanho mundanças nos valores mudam o formato
tamanho = 8 * escala
dme = 1 * escala
metade = tamanho / 2

#Coordenadas armazenadas em listas valores fixos e valores definidos atraves de operações entre as variaveis
xpontos = [0, metade - dme, metade, metade + dme, tamanho, metade + dme, metade, metade - dme, 0]
ypontos = [metade, metade + dme, tamanho, metade + dme, metade, metade - dme, 0, metade - dme, metade]

plt.plot(xpontos, ypontos)

plt.title("Desafio 01 Estrela com tamanho ajustavél")
plt.ylabel("Eixo Y")
plt.xlabel("Eixo x")

plt.grid(True)
plt.show()
```
